```python
from llvmlite import ir, binding

# Create an LLVM module and function
module = ir.Module(name="my_module")
func_type = ir.FunctionType(ir.IntType(32), [])
func = ir.Function(module, func_type, name="my_function")

# Create a basic block and an instruction builder
block = func.append_basic_block(name="entry")
builder = ir.IRBuilder(block)

# Build a simple function that returns 42
retval = ir.Constant(ir.IntType(32), 42)
builder.ret(retval)

# Print the LLVM IR
print(module)

# Initialize LLVM and JIT compile the module
binding.initialize()
binding.initialize_native_target()
binding.initialize_native_asmprinter()

target = binding.Target.from_default_triple()
target_machine = target.create_target_machine()
backing_mod = binding.parse_assembly(str(module))
engine = binding.create_mcjit_compiler(backing_mod, target_machine)

engine.finalize_object()
engine.run_static_constructors()

# Get the pointer to the function and call it
func_ptr = engine.get_function_address("my_function")
from ctypes import CFUNCTYPE, c_int
cfunc = CFUNCTYPE(c_int)(func_ptr)
result = cfunc()
print("Result from JITed function:", result)
```
